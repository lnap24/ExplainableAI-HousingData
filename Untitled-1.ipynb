```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(filtered_df):
    # Drop column: 'edit_status_name'
    filtered_df = filtered_df.drop(columns=['edit_status_name'])
    # Drop columns: 'sequence_number', 'edit_status' and 23 other columns
    filtered_df = filtered_df.drop(columns=['sequence_number', 'edit_status', 'rate_spread', 'denial_reason_3', 'denial_reason_name_3', 'denial_reason_2', 'denial_reason_name_2', 'denial_reason_1', 'denial_reason_name_1', 'co_applicant_race_5', 'co_applicant_race_name_5', 'co_applicant_race_4', 'co_applicant_race_name_4', 'co_applicant_race_3', 'co_applicant_race_name_3', 'co_applicant_race_2', 'co_applicant_race_name_2', 'applicant_race_5', 'applicant_race_name_5', 'applicant_race_4', 'applicant_race_name_4', 'applicant_race_3', 'applicant_race_name_3', 'applicant_race_2', 'applicant_race_name_2'])
    # Drop columns: 'state_code', 'state_abbr' and 5 other columns
    filtered_df = filtered_df.drop(columns=['state_code', 'state_abbr', 'state_name', 'agency_abbr', 'respondent_id', 'as_of_year', 'application_date_indicator'])
    return filtered_df

filtered_df_clean = clean_data(filtered_df.copy())
filtered_df_clean.head()
```

```python
# List of columns to one-hot encode
columns_to_encode = [
    'agency_name', 'loan_type_name', 'property_type_name', 'loan_purpose_name', 
    'owner_occupancy_name', 'preapproval_name', 'applicant_ethnicity_name', 
    'co_applicant_ethnicity_name', 'applicant_race_name_1', 'co_applicant_race_name_1', 
    'applicant_sex_name', 'co_applicant_sex_name', 'purchaser_type_name', 
    'hoepa_status_name', 'lien_status_name', 'msamd_name', 'county_name'
]

# One-hot encode the specified columns
encoded_df = pd.get_dummies(filtered_df_clean_1, columns=columns_to_encode)

# Display the first few rows of the encoded dataframe
encoded_df.head()
```

```python
# Drop the specified columns
encoded_df = encoded_df.drop(columns=['msamd', 'county_code'])

# Display the first few rows of the updated dataframe
encoded_df.head()
```

```python
print(encoded_df['action_taken_name'].unique())
```

1 = approved ['loan originated', 'loan purchased by the institution','Preapproval request approved but not accepted','application approved but not accepted']
2 = denied ['application denied by financial institution', 'Preapproval request denied by financial institution']
3 = Other ['file closed for incompleteness', 'application withdrawn by applicant'] 

```python
# Ensure action_taken_name is a category
encoded_df['action_taken_name'] = encoded_df['action_taken_name'].astype('category')

# Define the mapping dictionary
action_taken_map = {
    'Loan originated': 'approved',
    'Loan purchased by the institution': 'approved',
    'Preapproval request approved but not accepted': 'approved',
    'Application approved but not accepted': 'approved',
    'Application denied by financial institution': 'denied',
    'Preapproval request denied by financial institution': 'denied',
    'File closed for incompleteness': 'other',
    'Application withdrawn by applicant': 'other'
}

# Map the action_taken_name to the new categories
encoded_df['action_taken_name'] = encoded_df['action_taken_name'].map(action_taken_map)

# Handle unexpected values (if necessary)
encoded_df['action_taken_name'] = encoded_df['action_taken_name'].fillna('unknown')

# Display the first few rows of the updated dataframe
print(encoded_df[['action_taken_name']].head())
```
